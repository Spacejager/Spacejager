```python
import matplotlib 
import matplotlib.pyplot as plt
import numpy as np
print(matplotlib.__version__)
print(np.__version__)
```

Simple line plot

```python
x = np.arange(0.0, 2.0,0.01)
x
```

```python
s = np.sin(2 * np.pi * x)
s
```

```python
fig, ax = plt.subplots() #00 Style

ax.plot(x,s)
ax.set(xlabel='nilai x',
      ylabel='nilai sin(x)',
      title='visualisasi nilai sin')
ax.grid()
plt.show
```

Mutliple sub plots

```python
x1 = np.linspace(0.0,5.0,100)
x2 = np.linspace(0.0,2.0,100)

y1 = np.cos(2 * np.pi * x1)
y2 = np.cos(2 * np.pi * x2) 
```

```python
x1
```

Multiple Subplot dengan OO style

```python
fig,(ax1,ax2) =plt.subplots(2,1)
fig.suptitle('Multiple Subplots')

ax1.plot(x1,y1,'ro-')
ax1.set_ylabel('Nilai cos(X1)')

ax2.plot(x2,y2,'g.-')
ax1.set_ylabel('Nilai cos(X2)')
ax2.set_xlabel('Nilai x')
```

Multiple Subplot dengan pyplot Style

```python
plt.subplot(2,1,1)
plt.plot(x1,y1,'ro-')
plt.title('multiple subplots')
plt.ylabel('nilai $cos(x_1)$')

plt.subplot(2,1,2)
plt.plot(x2,y2,'g-')
plt.ylabel('nilai $cos(x_2)$')
plt.xlabel('Nilai $x$')

plt.show()
```

Histogram

Import Modules

```python
import matplotlib 
import matplotlib.pyplot as plt
import numpy as np
print(matplotlib.__version__)
print(np.__version__)
```

Sample Data

```python
mu,sigma =100,15 #nilai mean dan nilai standard deviation
x= mu + sigma * np.random.randn(10000)# normal distribution
x
```

```python
x.shape
```

Histogram dengan plypot style

```python
plt.hist(x,
        bins=50,
        facecolor='g',
        alpha=0.75)

plt.xlabel('sumbu x')
plt.ylabel('sumbu y')
plt.title('contoh histogram')

plt.text(45,500, r'\mu=100,\ \sigma=15$')
plt.grid()

plt.show()
```

Histogram dengan OO Style

```python
plt.hist(x,
        bins=50,
        facecolor='b',
        alpha=0.25)

plt.xlabel('sumbu x')
plt.ylabel('sumbu y')
plt.title('contoh histogram')

plt.text(45,500, r'\mu=100,\ \sigma=15$')
plt.grid()

plt.show()
```
