Box Plot/Box Chart

Dalam sesi ini kita belajar cara membuat box plot dengan Matplotlib

Import Modules

```python
%matplotlib inline
```

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

print(matplotlib.__version__)
print(np.__version__)
```

Sample Dataset

```python
np.random.seed(2)
data = np.random.normal(loc=100, scale=10, size=200)
data
```

Membuat Simple Plot

```python
plt.boxplot(data, labels=['Data'])

plt.title('Simple Box Plot')
plt.xlabel('Sumbu X')
plt.ylabel('Sumbu Y')

plt.grid()

plt.show()
```

Notched Box Plot

```python
plt.boxplot(data, labels=['Data'],notch=True)




plt.title('Notched Box Plot')
plt.xlabel('Sumbu X')
plt.ylabel('Sumbu Y')

plt.show()
```

Pengaturan maker untuk  outlier

```python
plt.boxplot(data,
           labels= ['Data'],
           showfliers=True,
           flierprops={'markerfacecolor' : 'r', 'marker': 'd'})
plt.title('Pengaturan Marker')
plt.xlabel('Sumbu X')
plt.ylabel('Sumbu Y')

plt.show()
```

Horizontal Box Plot

```python
plt.boxplot(data, labels=['Data'],vert=False)

plt.title('Horizontal Box Plot')
plt.xlabel('Sumbu X')
plt.ylabel('Sumbu Y')

plt.show()
```

Multiple Box Plot

```python

np.random.seed(10)
data1 = np.random.normal(100,10,200)
data2 = np.random.normal(80,30,200)
data3 = np.random.normal(90,20,200)
data4 = np.random.normal(70,25,200)

data = [data1,data2,data3,data4]
labels = ['Data1','Data2','Data3','Data4']
```

```python
plt.boxplot(data, labels=labels)

plt.title('Mutliple Box Plot')
plt.xlabel('Sumbu X')
plt.ylabel('Sumbu Y')


plt.show()
```

Twin Axes

Dalam Sesi ini kita belajar twin axes pada matplotlib.

```python
%matplotlib inline
```

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

print(matplotlib.__version__)
print(np.__version__)
```

Kasus Twin Axes 1

```python
x = np.linspace(1,10,25)
x
```

```python
fig, ax1 =plt.subplots()

ax1.plot(x, np.exp(x), 'bs-', label='exp')
ax1.set_xlabel('Sumbu X')
ax1.set_ylabel('exp')

ax2= ax1.twinx()
ax2.plot(x, np.log(x), 'ro-', label='log')
ax2.set_ylabel('log')


fig.suptitle('Contoh Twin Axes')
fig.legend(loc='upper left')

plt.show()
```

Kasus Twin Axes 2

```python
x = np.random.randint(16,40, size=30)
x
```

```python
def C2F(celcius=0):
    return (celcius * 1.8) + 32

def konversi_sumbu(ax1):
    y1, y2= ax1.get_ylim()
    ax2.set_ylim(C2F(y1), C2F(y2))
    ax2.figure.canvas.draw()
    
fig, ax1 = plt.subplots()
ax2 = ax1.twinx()
    
ax1.callbacks.connect('ylim_changed',konversi_sumbu)
ax1.plot(x)
    
ax1.set_xlabel('Hari')
ax1.set_ylabel('celcius')
ax2.set_ylabel('Fahrenheit')
    
fig.suptitle('Temperatur Udara')
plt.show()  
```
