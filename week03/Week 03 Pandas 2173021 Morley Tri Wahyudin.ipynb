#09 Membagi Data Frame Menjadi dua Secara Random

Import Modules

```python
import pandas as pd
import numpy as np
print(pd.__version__)
print(np.__version__)
```

```python
n_rows = 10
n_cols = 5
cols = tuple('abcde')

df= pd.DataFrame(np.random.randint(1,20, size=(n_rows,n_cols)),columns=cols)
df
```

Membagi Data frame menjadi dua secara acak berdasarkan proposi tertentu

```python
df.shape
```

```python
proposi=0.7
df_1 = df.sample(frac=proposi)
df_2 = df.drop(df_1.index)

print(f'df_1 shape: {df_1.shape}')
print(f'df_2 shape: {df_2.shape}')
```

#10: Mengganti nama(Label) kolom pada Data Frame berdasarkan pola

Import Modules

```python
import pandas as pd
print(pd.__version__)
```

Persiapan Data Frame

```python
df = pd.read_csv('titanic3.csv')
df.columns =['Pclass','Survived','name','Sex','age',
             'sibsp','Parch','Ticket','Fare','Cabin','Embarked','boat','body','home']
df_backup = df.copy(deep=True)

df.head()
```

menggunakan lowercase untuk nama kolom dan mengganti spasi dengan _

```python
df.columns.str.replace('','_').str.lower()
df.head()
```

Menangkas kelebihan spasi pada nama kolom

```python
df = df_backup.copy(deep=True)

df.columns =df.columns.str.lower().str.strip().str.replace('','_')
df.head()
```

#11: Melakukan Seleksi Kolom dan baris pada Data Frame menggunakan loc

Import Modules

```python
import pandas as pd
import numpy as np
print(pd.__version__)
print(np.__version__)
```

```python
n_rows = 10
n_cols = 5
cols = tuple('abcde')

df= pd.DataFrame(np.random.randint(1,20, size=(n_rows,n_cols)),columns=cols)
df
```

seleksi kolom dan baris dengan loc

```python
df.loc[[0,3,4], ['b','e']] 
```

Seleksi baris dengan condition

```python
df.loc[df['b']>10, ['b','d','e']]
```

Slicing Data Frame dengan loc

```python
df.loc[0:4, 'b':'d']
```

#12: Membuat kolom tipe datatime dari sejumlah kolom lain di Data Frame

```python
import pandas as pd
print(pd.__version__)
```

Persiapan Data Frame

```python
data ={'day':[1,2,10,25,12],
      'month':[1,2,4,5,6],
      'year':[2000,2001,2010,2015,2020]}
df = pd.DataFrame(data)
df
```

membentuk kolom bertipe datatime

```python
df['penaggalan'] = pd.to_datetime(df[['day','month','year']])
df
```
