Bar plot/Bar Chart

dalam sesi ini kita mempelajari membuat bar plot dengan matplotlib

Import Modules

```python
%matplotlib inline
```

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
print(matplotlib.__version__)
print(np.__version__)
```

Simple Bar Plot

```python
data = [25,45,55,125,225]
kategori = ['a','b','c','d','e']
```

```python
plt.bar(kategori, data)
plt.grid()

plt.xlabel('Kategori')
plt.ylabel('jumlah')
plt.title('contoh bar plot')

plt.show()
```

Pengaturan Grid dan Color

```python
plt.bar(kategori,
        data,
        color='red',
       alpha=0.75)
plt.grid(linestyle='--',
        linewidth=4,
        axis='y',#ini line untuk bagian mana yang diwarnai
        alpha=0.50)

plt.xlabel('Kategori')
plt.ylabel('jumlah')
plt.title('contoh bar plot')

plt.show()
```

Horizontal Bar Plot

```python
data = [25,45,55,125,225]
kategori = ['a','b','c','d','e']
```

```python
plt.barh(kategori,data)
plt.grid(linestyle='--',
        linewidth=1,
        axis='x',#ini line untuk bagian mana yang diwarnai
        alpha=0.75)

plt.xlabel('Kategori')
plt.ylabel('jumlah')
plt.title('contoh bar plot')

plt.show()

```

Stacked Bar Plot

```python
data1 =[25,85,75,40,60]
data2 =[40,35,22,55,11]
kategori = ['a','b','c','d','e']
plt.bar(kategori,data1, label='Data 1')
plt.bar(kategori,data1, label='Data 2',bottom=data1)
plt.legend()

plt.grid(linestyle='--',
        linewidth=1,
        axis='y',
        alpha=0.75)

plt.xlabel('Kategori')
plt.ylabel('jumlah')
plt.title('contoh bar plot')

plt.show()
```

Grouped Bar Plot

```python
data1 =[25,85,75,40,60]
data2 =[40,35,22,55,11]
kategori = ['a','b','c','d','e']

x= np.arange(len(kategori))
x
```

```python
width= 0.25

plt.bar(x-width/2,data1,width,label='Data1')
plt.bar(x+width/2,data1,width,label='Data2')

plt.xticks(x,kategori)

plt.grid(linestyle='--',
        linewidth=1,
        axis='y',
        alpha=0.75)

plt.xlabel('Kategori')
plt.ylabel('jumlah')
plt.title('contoh Grouped bar plot')

plt.legend()
plt.show()

```

Pie plot/pie chart

kita akan belajar membuat pie plot dengan matplotlib.

```python
%matplotlib inline
```

```python
import matplotlib
import matplotlib.pyplot as plt

print(matplotlib.__version__)
```

Simple Pie plot

```python
kategori = ['a','b','c','d','e']
data = [215,130,245,210,225]
```

```python
plt.pie(data,
       labels=kategori,
       autopct='%1.1f%%',
       startangle=90)
plt.title('contoh pie plot')
plt.show()
```

Pengaturan Legend dan color

```python
kategori = ['a','b','c','d','e']
data = [215,130,245,210,225]
warna = ['pink','cyan','#e03364','yellowgreen','skyblue']
```

```python
plt.pie(data,
       labels=kategori,
       colors=warna,
       autopct='%1.1f%%',
       startangle=90)
plt.legend()
plt.title('contoh pie plot')
plt.show()
```

Exploded Pie Plot

```python
kategori = ['a','b','c','d','e']
data = [215,130,245,210,225]
warna = ['pink','cyan','#e03364','yellowgreen','skyblue']
explode_var = [0.,0.,0.2,0.,0.08]
```

```python
plt.pie(data,
       labels=kategori,
       colors=warna,
       explode=explode_var,
       autopct='%1.1f%%',
       startangle=90)

plt.title('contoh pie plot')
plt.show()
```

Table Plot/Table Chart

Kita akan belajar membuat table plot dengan matplotlib

```python
%matplotlib inline
```

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
print(matplotlib.__version__)
print(np.__version__)
```

Sample Data

```python
data_uas =[
    ['bejo',70],
    ['tesso',83],
    ['cecep',62],
    ['warren',74],
    ['karen',71]
]
```

```python
data_uas
```

Table Plot Dengan oo Style

```python
fig, ax= plt.subplots()

table = plt.table(cellText=data_uas,
                 loc='center')

table.set_fontsize(14)
table.scale(1,4)#size kolom,baris
ax.axis(False)


plt.show()
```

Table Plot dengan pyplot Style

```python
table= plt.table(cellText=data_uas,
                loc='center')

table.set_fontsize(14)
table.scale(1,4)#size kolom,baris

ax=plt.gca()
ax.axis(False)

plt.show()
```
